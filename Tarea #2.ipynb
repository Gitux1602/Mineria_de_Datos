```python
import pandas as pd # Importamos la libreria pandas
```

```python
df = pd.read_csv('https://www.football-data.co.uk/new/ARG.csv') # Leemos el archivo csv
```

```python
df['Date'] = pd.to_datetime(df['Date']) # Cambiamos la columna 'Date' en un formato de fecha y hora adecuado que Pandas pueda reconocer
```

```python
boca_river_matches = df[(df['Home'] == 'Boca Juniors') & (df['Away'] == 'River Plate') | (df['Home'] == 'River Plate') & (df['Away'] == 'Boca Juniors')]
# Guarda todos los partidos entre Boca Juniors y River Plate //Los equipos mas grandes de Argentina avalado por sus t√≠tulos 
```

```python
boca_river_matches[boca_river_matches['Date'].dt.year >= 2015] # Muestra los partidos entre boca y river del a√±o 2015 en adelante
```

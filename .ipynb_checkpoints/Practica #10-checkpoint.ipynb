```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def cargar_datos(url):
    # Cargar el CSV desde la URL
    df = pd.read_csv(url)
    return df

def calcular_desempeno_equipo(df, equipo):
    # Filtrar partidos donde el equipo es local o visitante
    partidos_local = df[df['Home'] == equipo]
    partidos_visitante = df[df['Away'] == equipo]

    # Calcular el desempeño en casa y fuera
    desempeno_local = partidos_local['Res'].value_counts()
    desempeno_visitante = partidos_visitante['Res'].value_counts()

    return desempeno_local, desempeno_visitante

def generar_mapa_calor(desempeno_local, desempeno_visitante, equipo):
    # Crear un DataFrame para el mapa de calor
    mapa_calor_data = pd.DataFrame({'Local': desempeno_local, 'Visitante': desempeno_visitante}).T.fillna(0)

    # Normalizar los valores para obtener proporciones
    mapa_calor_data = mapa_calor_data.div(mapa_calor_data.sum(axis=1), axis=0)

    # Crear el mapa de calor con seaborn
    sns.heatmap(mapa_calor_data, annot=True, cmap='Blues', fmt=".2%")
    
    plt.title(f'Desempeño de {equipo} en Casa/Fuera')
    plt.xlabel('Resultado')
    plt.ylabel('Local/Visitante')
    plt.show()

def main():
    # URL del CSV
    url = "https://www.football-data.co.uk/new/ARG.csv"
    
    # Cargar datos
    df = cargar_datos(url)
    
    # Obtener la lista única de equipos
    equipos = df['Home'].unique()

    # Solicitar al usuario el equipo a analizar
    equipo_seleccionado = input(f"Seleccione un equipo ({', '.join(equipos)}): ")

    # Verificar si el equipo seleccionado está en la lista
    if equipo_seleccionado not in equipos:
        print("Equipo no válido. Por favor, elija un equipo válido.")
        return

    # Calcular el desempeño del equipo
    desempeno_local, desempeno_visitante = calcular_desempeno_equipo(df, equipo_seleccionado)

    # Generar y mostrar el mapa de calor
    generar_mapa_calor(desempeno_local, desempeno_visitante, equipo_seleccionado)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd # Importamos la libreria pandas
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('https://www.football-data.co.uk/new/ARG.csv') # Leemos el archivo csv
```

```python
boca_river_matches = df[(df['Home'] == 'Boca Juniors') & (df['Away'] == 'River Plate') | (df['Home'] == 'River Plate') & (df['Away'] == 'Boca Juniors')]
# Guarda todos los partidos entre Boca Juniors y River Plate //Los equipos mas grandes de Argentina avalado por sus títulos 
```

```python
boca_river_matches
```

```python
# Filtrar los partidos de Boca Juniors y River Plate cuando juegan como equipo local o visitante
boca_local = df[(df['Home'] == 'Boca Juniors') & (df['Away'] == 'River Plate')]
boca_visitante = df[(df['Away'] == 'Boca Juniors') & (df['Home'] == 'River Plate')]
river_local = df[(df['Home'] == 'River Plate') & (df['Away'] == 'Boca Juniors')]
river_visitante = df[(df['Away'] == 'River Plate') & (df['Home'] == 'Boca Juniors')]
```

```python
# Calcular estadísticas para Boca Juniors como equipo local
media_boca_local_HG = boca_local['HG'].mean()
moda_boca_local_HG = boca_local['HG'].mode()[0]
mediana_boca_local_HG = boca_local['HG'].median()
varianza_boca_local_HG = boca_local['HG'].var()
print("Estadísticas para Boca Juniors como equipo local:")
print("Media HG:", media_boca_local_HG)
print("Moda HG:", moda_boca_local_HG)
print("Mediana HG:", mediana_boca_local_HG)
print("Varianza HG:", varianza_boca_local_HG)
```

```python
# Calcular estadísticas para Boca Juniors como equipo visitante
media_boca_visitante_AG = boca_visitante['AG'].mean()
moda_boca_visitante_AG = boca_visitante['AG'].mode()[0]
mediana_boca_visitante_AG = boca_visitante['AG'].median()
varianza_boca_visitante_AG = boca_visitante['AG'].var()
print("\nEstadísticas para Boca Juniors como equipo visitante:")
print("Media AG:", media_boca_visitante_AG)
print("Moda AG:", moda_boca_visitante_AG)
print("Mediana AG:", mediana_boca_visitante_AG)
print("Varianza AG:", varianza_boca_visitante_AG)
```

```python
# Calcular estadísticas para River Plate como equipo local
media_river_local_HG = river_local['HG'].mean()
moda_river_local_HG = river_local['HG'].mode()[0]
mediana_river_local_HG = river_local['HG'].median()
varianza_river_local_HG = river_local['HG'].var()
print("\nEstadísticas para River Plate como equipo local:")
print("Media HG:", media_river_local_HG)
print("Moda HG:", moda_river_local_HG)
print("Mediana HG:", mediana_river_local_HG)
print("Varianza HG:", varianza_river_local_HG)
```

```python
# Calcular estadísticas para River Plate como equipo visitante
media_river_visitante_AG = river_visitante['AG'].mean()
moda_river_visitante_AG = river_visitante['AG'].mode()[0]
mediana_river_visitante_AG = river_visitante['AG'].median()
varianza_river_visitante_AG = river_visitante['AG'].var()
print("\nEstadísticas para River Plate como equipo visitante:")
print("Media AG:", media_river_visitante_AG)
print("Moda AG:", moda_river_visitante_AG)
print("Mediana AG:", mediana_river_visitante_AG)
print("Varianza AG:", varianza_river_visitante_AG)
```

```python
#P4
# Creamos una variable 'goles_boca_river' que contiene los datos de goles de los partidos entre Boca Juniors y River Plate.
goles_boca_river = boca_river_matches.copy()

# Crea un gráfico de boxplot para visualizar los goles anotados por Boca Juniors y River Plate, ya sea como equipo local o visitante.
plt.figure(figsize=(8, 6))
plt.boxplot(
            [
             goles_boca_river[(goles_boca_river['Home'] == 'Boca Juniors') & (goles_boca_river['Away'] == 'River Plate')]['HG'],
             goles_boca_river[(goles_boca_river['Home'] == 'River Plate') & (goles_boca_river['Away'] == 'Boca Juniors')]['HG'],
             goles_boca_river[(goles_boca_river['Away'] == 'Boca Juniors') & (goles_boca_river['Home'] == 'River Plate')]['AG'],
             goles_boca_river[(goles_boca_river['Away'] == 'River Plate') & (goles_boca_river['Home'] == 'Boca Juniors')]['AG']
            ],
            labels=['Boca Juniors(Local)', 'River Plate(Local)', 'Boca Juniors(Visitante)', 'River Plate(Visitante)']
           )

plt.title('Goles Anotados por Boca Juniors y River Plate (Local y Visitante)')
plt.xlabel('Equipo')
plt.ylabel('Goles Anotados')

# Muestra el gráfico.
plt.show()
```
